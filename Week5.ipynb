# Plotting Using Matplotlib

1. Plotting simple functions, labels, legends, multiple plots


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot the sine function
plt.plot(x, y, label='sin(x)')  # Plot the data
plt.xlabel('x')  # Label for x-axis
plt.ylabel('y')  # Label for y-axis
plt.title('Simple Sin Function')  # Title for the plot
plt.legend()  # Show legend
plt.show()  # Display the plot
```

2. About scatter plots


```python
# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 100, size=50)  # Random sizes for markers

# Create scatter plot
plt.scatter(x, y, s=sizes, alpha=0.5)  # Plot the data points
plt.xlabel('X')  # Label for x-axis
plt.ylabel('Y')  # Label for y-axis
plt.title('Scatter Plot Example')  # Title for the plot
plt.show()  # Display the plot
```

3. Bar chart


```python
# Data for bar chart
x = ['A', 'B', 'C', 'D', 'E']
y = [10, 15, 7, 10, 12]

# Create bar chart
plt.bar(x, y)  # Plot the bars
plt.xlabel('Categories')  # Label for x-axis
plt.ylabel('Values')  # Label for y-axis
plt.title('Bar Chart Example')  # Title for the plot
plt.show()  # Display the plot
```

4. Histogram

```python
# Generate random data
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, edgecolor='black')  # Plot histogram
plt.xlabel('Value')  # Label for x-axis
plt.ylabel('Frequency')  # Label for y-axis
plt.title('Histogram Example')  # Title for the plot
plt.show()  # Display the plot
```

5. Pie chart


```python
# Data for pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [155, 302, 458, 104]

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.2f%%')  # Plot pie chart
plt.title('Pie Chart Example')  # Title for the plot
plt.show()  # Display the plot
```

6. Changing styles of plots


```python
# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Plot with a custom style
plt.style.use('ggplot')  # Set the style
plt.plot(x, y)  # Plot the data
plt.xlabel('X')  # Label for x-axis
plt.ylabel('Y')  # Label for y-axis
plt.title('Plot with Custom Style')  # Title for the plot
plt.show()  # Display the plot
```

#Simple Linear Regression

1. Importing Needed packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

2. Downloading data

```python
!wget -O FuelConsumption.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/FuelConsumptionCo2.csv
```

3. Understanding the Data

```python
df = pd.read_csv("FuelConsumption.csv")

# take a look at the dataset
df.head(5)
```

4. Data Exploration

```python
# summarize the data
df.describe()
```

5. Lets select some features to explore more.

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

6. we can plot each of these features:

```python
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

7. Now, lets plot each of these features vs the Emission, to see how linear is their relation:

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
# write your code here
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='blue')
plt.xlabel("Cylinders")
plt.ylabel("Emission")
plt.show()
```

8. Creating train and test dataset

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
print(len(test))
```

9. Train data distribution

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

10. Modeling

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

11. Plot outputs

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

12. Evaluation

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
